```python
import pandas as pd

toxic_response_factors = {
    'Cd': 30,  # Highest toxicity - Hakanson (1980)
    'Pb': 5,   # Moderate toxicity - Hakanson (1980)
    'Cu': 5,   # Moderate toxicity - Hakanson (1980)
    'Cr': 2,   # Low toxicity - Hakanson (1980)
    'Ni': 5,   # Moderate toxicity - recent studies
    'Co': 5,   # Moderate toxicity - recent studies
    'Zn': 1,   # Lowest toxicity - Hakanson (1980)
    'Fe': 1,   # Essential element - default
    'Mn': 1    # Essential element - default
}

def calculate_mcd():
    # Read CF results
    cf = pd.read_csv("Ajakanga_output/CF_results.csv")
    
    # Get metals list (excluding sample_location column)
    metals = cf.columns[1:]
    
    # Initialize lists to store results for each sample
    mcd_values = []
    
    # Calculate MCD for each sample (row)
    for idx, row in cf.iterrows():
        weighted_sum = 0
        total_tr = 0
        
        for metal in metals:
            cf_value = row[metal]
            
            if metal in toxic_response_factors:
                tr = toxic_response_factors[metal]
                weighted_sum += cf_value * tr
                total_tr += tr
            else:
                # If metal not in dictionary, use default Tr = 1
                print(f"Warning: {metal} not found in toxic response factors. Using default Tr=1")
                weighted_sum += cf_value * 1
                total_tr += 1
        
        # Calculate MCD for this sample = Σ(CF × Tr) / Σ Tr
        if total_tr > 0:
            mcd = weighted_sum / total_tr
        else:
            mcd = 0
            print(f"Warning: No valid Tr values for sample {row['sample_location']}")
        
        mcd_values.append(mcd)
    
    # Create result dataframe
    result = pd.DataFrame({
        "Sample": cf["Sample"],
        "mCd": mcd_values
    })
    
    # Add classification
    def classify_mcd(value):
        if value < 1.0:
            return "Low Contamination"
        elif 1.0 <= value < 2.0:
            return "Moderate Contamination"
        elif 2.0 <= value < 3.5:
            return "Considerable Contamination"
        else:
            return "High Contamination"
    
    result["Classification"] = result["mCd"].apply(classify_mcd)
    
    # Save results
    result.to_csv("Ajakanga_output/mCD_results.csv", index=False)
    print(f"✓ MCD calculations complete. Processed {len(result)} samples")
    
    return result
```

```python
import pandas as pd
import numpy as np

def classify_pli(pli_value):
    """
    Classify a single Pollution Load Index (PLI) value.
    """
    if pli_value < 1.0:
        return "Unpolluted"
    elif 1.0 <= pli_value < 2.0:
        return "Slightly Polluted"
    elif 2.0 <= pli_value < 3.0:
        return "Moderately Polluted"
    elif 3.0 <= pli_value < 4.0:
        return "Highly Polluted"
    else:
        return "Extremely Polluted"

def calculate_pli():
    # 1. Load Contamination Factors
    cf = pd.read_csv("Ajakanga_output/CF_results.csv")

    # 2. Identify metal columns (all columns except the first, assumed to be 'Sample')
    metals = cf.columns[1:]

    # 3. Compute PLI for each sample (geometric mean of CFs)
    pli = (cf[metals].prod(axis=1)) ** (1 / len(metals))

    # 4. Build result DataFrame
    result = pd.DataFrame({
        "Sample": cf.iloc[:, 0],          # first column = sample names
        "PLI": pli
    })

    # 5. **Add interpretation layer** – classify each PLI
    result["Classification"] = result["PLI"].apply(classify_pli)

    # 6. Save to CSV
    result.to_csv("Ajakanga_output/PLI_results.csv", index=False)

    return result

if __name__ == "__main__":
    calculate_pli()
```

```python
import pandas as pd
import numpy as np

def calculate_npi():

    data = pd.read_csv("AJAKANGAHM.csv")

    perm = permissible

    metals = data.columns[1:]

    ratios = data.copy()

    for metal in metals:
        ratios[metal] = data[metal] / perm[metal]

    mean_ratio = ratios[metals].mean(axis=1)
    max_ratio = ratios[metals].max(axis=1)

    npi = np.sqrt((mean_ratio**2 + max_ratio**2)/2)

    result = pd.DataFrame({
        "Sample": data.Sample,
        "NPI": npi
    })

    result.to_csv("Ajakanga_output/NPI_results.csv", index=False)

    return result

```

```python
import pandas as pd
import numpy as np

def classify_igeo(val):

    if val <= 0:
        return "Unpolluted","Background concentration."
    elif val <= 1:
        return "Unpolluted to Moderately Polluted","Minor contamination."
    elif val <= 2:
        return "Moderately Polluted","Moderate enrichment."
    elif val <= 3:
        return "Moderately to Heavily Polluted","Significant contamination."
    elif val <= 4:
        return "Heavily Polluted","Serious anthropogenic impact."
    elif val <= 5:
        return "Heavily to Extremely Polluted","Very high pollution."
    else:
        return "Extremely Polluted","Severe contamination."


def calculate_igeo():

    data = pd.read_csv("AJAKANGAHM.csv")
    background = pd.read_csv("standard_limits.csv")

    bg = dict(zip(background.Metal,background.Background))
    metals = data.columns[1:]

    results = []

    for i,row in data.iterrows():

        for metal in metals:

            val = np.log2(row[metal] / (1.5 * bg.get(metal,1)))

            c,i2 = classify_igeo(val)

            results.append([
                row["Sample"],
                metal,
                val,
                c,
                i2
            ])

    result = pd.DataFrame(results,columns=[
        "Sample","Metal","Igeo","Classification","Interpretation"
    ])

    result.to_csv("Ajakanga_output/IGEO_results.csv",index=False)

    return result
```

```python
import pandas as pd
import numpy as np

def classify_hpi(value):

    if value < 50:
        return "Excellent","Water is safe with insignificant heavy metal pollution."
    elif value < 100:
        return "Good","Water is slightly affected but still within acceptable limits."
    elif value < 300:
        return "Poor","Water shows considerable heavy metal contamination."
    else:
        return "Very Poor","Water is critically polluted and unsafe."


def calculate_hpi():

    data = pd.read_csv("AJAKANGAHM.csv")
    limits = pd.read_csv("standard_limits.csv")

    # Clean column names (VERY IMPORTANT)
    data.columns = data.columns.str.strip()
    limits.columns = limits.columns.str.strip()

    # Dictionaries
    perm = dict(zip(limits["Metal"], limits["Permissible"]))

    # Handle Ideal safely
    if "Ideal" in limits.columns:
        ideal = dict(zip(limits["Metal"], limits["Ideal"]))
    else:
        ideal = {metal: 0 for metal in limits["Metal"]}

    metals = data.columns[1:]

    weights = {}
    valid_metals = []

    # Compute weights safely
    for metal in metals:
        Si = perm.get(metal, np.nan)

        if pd.notna(Si) and Si != 0:
            weights[metal] = 1 / Si
            valid_metals.append(metal)

    # If no valid metals → stop
    if len(valid_metals) == 0:
        raise ValueError("No valid metals found. Check your limits file.")

    sub_index = pd.DataFrame(index=data.index)
    weighted_sub_index = pd.DataFrame(index=data.index)

    for metal in valid_metals:
        Mi = data[metal]
        Si = perm[metal]
        Ii = ideal.get(metal, 0)

        # Prevent division by zero
        denominator = (Si - Ii)
        if denominator == 0:
            continue

        Qi = ((Mi - Ii) / denominator) * 100

        # Replace inf and NaN
        Qi = Qi.replace([np.inf, -np.inf], np.nan).fillna(0)

        sub_index[metal] = Qi
        weighted_sub_index[metal] = Qi * weights[metal]

    sum_wqi = weighted_sub_index.sum(axis=1)
    sum_wi = sum(weights[m] for m in valid_metals)

    # Final HPI (safe division)
    hpi = sum_wqi / sum_wi if sum_wi != 0 else np.nan

    # Clean final result
    hpi = hpi.replace([np.inf, -np.inf], np.nan).fillna(0)

    classification = []
    interpretation = []

    for val in hpi:
        c, i = classify_hpi(val)
        classification.append(c)
        interpretation.append(i)

    result = pd.DataFrame({
        "Sample": data["Sample"],
        "HPI": hpi.round(3),
        "Classification": classification,
        "Interpretation": interpretation
    })

    result.to_csv("Ajakanga_output/HPI_results.csv", index=False)

    return result


if __name__ == "__main__":
    print(calculate_hpi())
```
