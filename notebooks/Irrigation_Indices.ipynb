```python
import pandas as pd
import numpy as np

def load_data():

    data = pd.read_csv("Aja_cat_an.csv")

    return data
```

```python
data = load_data()
```

```python
def calculate_sar(data):

    sar = data["Na"] / np.sqrt((data["Ca"] + data["Mg"]) / 2)

    classification = []
    interpretation = []

    for val in sar:

        if val < 10:
            classification.append("Excellent")
            interpretation.append("Suitable for irrigation with minimal sodium hazard")

        elif val < 18:
            classification.append("Good")
            interpretation.append("Suitable for most soils with moderate management")

        elif val < 26:
            classification.append("Doubtful")
            interpretation.append("May cause soil permeability problems")

        else:
            classification.append("Unsuitable")
            interpretation.append("Severe sodium hazard")

    result = pd.DataFrame({
        "Sample": data.Sample,
        "SAR": sar.round(3),
        "Class": classification,
        "Interpretation": interpretation
    })

    result.to_csv("Ajakanga_output/SAR_results.csv", index=False)

    return result
```

```python
calculate_sar(data)
```

```python
def calculate_na_percent(data):

    na_percent = ((data["Na"] + data["K"]) /
                 (data["Ca"] + data["Mg"] + data["Na"] + data["K"])) * 100

    classes=[]
    interp=[]

    for val in na_percent:

        if val < 20:
            classes.append("Excellent")
            interp.append("Safe irrigation water")

        elif val < 40:
            classes.append("Good")
            interp.append("Suitable for irrigation")

        elif val < 60:
            classes.append("Permissible")
            interp.append("Acceptable with moderate soil management")

        elif val < 80:
            classes.append("Doubtful")
            interp.append("Potential sodium hazard")

        else:
            classes.append("Unsuitable")
            interp.append("Not recommended for irrigation")

    result = pd.DataFrame({
        "Sample": data.Sample,
        "Na%": na_percent.round(2),
        "Class": classes,
        "Interpretation": interp
    })

    result.to_csv("Ajakanga_output/NaPercent_results.csv", index=False)

    return result
```

```python
calculate_na_percent(data)
```

```python
def calculate_kr(data):

    kr = data["Na"] / (data["Ca"] + data["Mg"])

    classification=[]
    interp=[]

    for val in kr:

        if val < 1:
            classification.append("Suitable")
            interp.append("Water suitable for irrigation")

        else:
            classification.append("Unsuitable")
            interp.append("Excess sodium may affect soil permeability")

    result = pd.DataFrame({
        "Sample": data.Sample,
        "KR": kr.round(3),
        "Class": classification,
        "Interpretation": interp
    })

    result.to_csv("Ajakanga_output/KR_results.csv", index=False)

    return result
```

```python
calculate_kr(data)
```

```python
#residual sodium carbonate
def calculate_rsc(data):

    rsc = (data["CO3"] + data["HCO3"]) - (data["Ca"] + data["Mg"])

    classes=[]
    interp=[]

    for val in rsc:

        if val < 1.25:
            classes.append("Safe")
            interp.append("Suitable for irrigation")

        elif val < 2.5:
            classes.append("Marginal")
            interp.append("May require soil management")

        else:
            classes.append("Unsuitable")
            interp.append("High alkalinity hazard")

    result = pd.DataFrame({
        "Sample": data.Sample,
        "RSC": rsc.round(3),
        "Class": classes,
        "Interpretation": interp
    })

    result.to_csv("Ajakanga_output/RSC_results.csv", index=False)

    return result
```

```python
calculate_rsc(data)
```

```python
def calculate_mh(data):

    mh = (data["Mg"] / (data["Ca"] + data["Mg"])) * 100

    classes=[]
    interp=[]

    for val in mh:

        if val < 50:
            classes.append("Suitable")
            interp.append("Magnesium hazard is low")

        else:
            classes.append("Unsuitable")
            interp.append("High magnesium hazard affecting soil quality")

    result = pd.DataFrame({
        "Sample": data.Sample,
        "MH": mh.round(2),
        "Class": classes,
        "Interpretation": interp
    })

    result.to_csv("Ajakanga_output/MH_results.csv", index=False)

    return result
```

```python
calculate_mh(data)
```

```python
weights = {
"Ca":2,
"Mg":2,
"Na":3,
"K":1,
"HCO3":2,
"CO3":2,
"Cl":3,
"SO4":3
}

standards = {
"Ca":75,
"Mg":50,
"Na":200,
"K":12,
"HCO3":300,
"CO3":75,
"Cl":250,
"SO4":250
}
```

```python
def calculate_wqi(data):

    parameters=list(weights.keys())

    qi=[]
    wi=[]

    for p in parameters:

        qi.append((data[p] / standards[p]) * 100)
        wi.append(weights[p])

    qi_df=pd.concat(qi,axis=1)
    qi_df.columns=parameters

    wqi=(qi_df * wi).sum(axis=1) / sum(wi)

    classes=[]
    interp=[]

    for val in wqi:

        if val < 50:
            classes.append("Excellent")
            interp.append("Water suitable for drinking and irrigation")

        elif val < 100:
            classes.append("Good")
            interp.append("Minor quality concerns")

        elif val < 200:
            classes.append("Poor")
            interp.append("Treatment recommended")

        elif val < 300:
            classes.append("Very Poor")
            interp.append("Limited suitability")

        else:
            classes.append("Unsuitable")
            interp.append("Not safe for consumption")

    result=pd.DataFrame({
        "Sample":data.Sample,
        "WQI":wqi.round(2),
        "Class":classes,
        "Interpretation":interp
    })

    result.to_csv("Ajakanga_output/WQI_results.csv",index=False)

    return result
```

```python
calculate_wqi(data)
```
