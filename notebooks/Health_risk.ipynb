```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("AJAKANGAHM.csv")

sample_col = "Sample"
```

```python
metals = ["Fe","Pb", "Cu","Cr", "Mn", "Zn", "Ni", "Co", "Cd"]
```

```python
# ============================================================
# CHILDREN EXPOSURE PARAMETERS
# Based on USEPA Exposure Factors Handbook
# ============================================================

# ------------------------------------------------------------
# Exposure parameters
# ------------------------------------------------------------

IR = 1.0          # ingestion rate (L/day)
EF = 365          # exposure frequency (days/year)
ED = 6            # exposure duration (years)
BW = 15           # body weight (kg)

AT_nc = ED * 365  # averaging time non-cancer
AT_c = 70 * 365   # averaging time cancer


# ------------------------------------------------------------
# Dermal exposure parameters
# ------------------------------------------------------------

SA = 6600         # exposed skin area (cm2)
Kp = 0.001        # dermal permeability coefficient (cm/hr)
ET = 1.0          # exposure time (hr/day)
CF = 0.001        # unit conversion factor
ABS = 0.001       # dermal absorption factor


# ------------------------------------------------------------
# Inhalation exposure parameters
# ------------------------------------------------------------

InhR = 7.6        # inhalation rate (m3/day)
PEF = 1.36e9      # particle emission factor


# ============================================================
# ORAL CANCER SLOPE FACTORS (CSF)
# Units: (mg/kg-day)^-1
# ============================================================

INGEST_SF = {
    'Fe': 0,
    'Pb': 0.0085,
    'Cu': 0,
    'Cr': 0.5,
    'Mn': 0,
    'Zn': 0,
    'Ni': 0.84,
    'Co': 0,
    'Cd': 0.38
}


# ============================================================
# DERMAL CANCER SLOPE FACTORS
# ============================================================

DERMAL_CSF = {
    'Fe': 0,
    'Pb': 0.017,
    'Cu': 0,
    'Cr': 20.0,
    'Mn': 0,
    'Zn': 0,
    'Ni': 22.75,
    'Co': 0,
    'Cd': 7.6
}


# ============================================================
# INHALATION CANCER SLOPE FACTORS
# ============================================================

INHALATION_CSF = {
    'Fe': 0,
    'Pb': 0.042,
    'Cu': 0,
    'Cr': 42.0,
    'Mn': 0,
    'Zn': 0,
    'Ni': 0.84,
    'Co': 9.8,
    'Cd': 6.3
}


# ============================================================
# INHALATION UNIT RISK
# ============================================================

INHALATION_IUR = {
    'Fe': 0,
    'Pb': 0.000012,
    'Cu': 0,
    'Cr': 0.012,
    'Mn': 0,
    'Zn': 0,
    'Ni': 0.00026,
    'Co': 0.009,
    'Cd': 0.0018
}


# ============================================================
# ORAL REFERENCE DOSE (RfD)
# Units: mg/kg-day
# ============================================================

ORAL_RfD = {
    'Fe': 0.7,
    'Pb': 0.0035,
    'Cu': 0.04,
    'Cr': 0.003,
    'Mn': 0.14,
    'Zn': 0.3,
    'Ni': 0.02,
    'Co': 0.02,
    'Cd': 0.001
}


# ============================================================
# DERMAL REFERENCE DOSE
# ============================================================

DERMAL_RfD = {
    'Fe': 0.14,
    'Pb': 0.000525,
    'Cu': 0.008,
    'Cr': 0.00006,
    'Mn': 0.00185,
    'Zn': 0.06,
    'Ni': 0.0008,
    'Co': 0.016,
    'Cd': 0.00001
}


# ============================================================
# INHALATION REFERENCE DOSE
# ============================================================

INHALATION_RfD = {
    'Fe': 0.00022,
    'Pb': 0.0035,
    'Cu': 0.04,
    'Cr': 0.0002,
    'Mn': 0.0000143,
    'Zn': 0.3,
    'Ni': 0.002,
    'Co': 0.00003,
    'Cd': 0.001
}
```

```python
def calculate_cdi_ingestion_child(df):

    results = []

    for metal in metals:

        if metal in df.columns:

            CDI = (df[metal] * IR * EF * ED) / (BW * AT_nc)

            for i,val in enumerate(CDI):

                results.append([
                    df[sample_col].iloc[i],
                    metal,
                    val
                ])

    table_ingestion = pd.DataFrame(results,
        columns=["Sample","Metal","CDI_Ingestion_Child"])

    table_ingestion.to_csv("Ajakanga_output\\Child_risk\\CDI_Ingestion_Child.csv",index=False)

    return table_ingestion
```

```python
def calculate_cdi_dermal_child(df):

    results = []

    for metal in metals:

        if metal in df.columns:

            CDI = (
                df[metal] * SA * Kp * ET * EF * ED * CF
            ) / (BW * AT_nc)

            for i,val in enumerate(CDI):

                results.append([
                    df[sample_col].iloc[i],
                    metal,
                    val
                ])

    table_dermal = pd.DataFrame(results,
        columns=["Sample","Metal","CDI_Dermal_Child"])

    table_dermal.to_csv("Ajakanga_output\\Child_risk\\CDI_Dermal_Child.csv",index=False)

    return table_dermal
```

```python
def calculate_cdi_inhalation_child(df):

    results = []

    for metal in metals:

        if metal in df.columns:

            CDI = (
                df[metal] * InhR * EF * ED
            ) / (PEF * BW * AT_nc)

            for i,val in enumerate(CDI):

                results.append([
                    df[sample_col].iloc[i],
                    metal,
                    val
                ])

    table_inhale = pd.DataFrame(results,
        columns=["Sample","Metal","CDI_Inhalation_Child"])

    table_inhale.to_csv("Ajakanga_output\\Child_risk\\CDI_Inhalation_Child.csv",index=False)

    return table_inhale
```

```python
def calculate_hq_ingestion(cdi_table, RfD):

    results = []

    for i,row in cdi_table.iterrows():

        metal = row["Metal"]

        if metal in RfD:

            HQ = row["CDI_Ingestion_Child"] / RfD[metal]

            results.append([
                row["Sample"],
                metal,
                row["CDI_Ingestion_Child"],
                HQ
            ])

    table = pd.DataFrame(results,
        columns=["Sample","Metal","CDI","HQ"])

    table.to_csv("Ajakanga_output\\Child_risk\\HQ_Ingestion_Child.csv",index=False)

    return table
```

```python
def calculate_hq_dermal(cdi_table, RfD):

    results = []

    for i,row in cdi_table.iterrows():

        metal = row["Metal"]

        if metal in RfD:

            HQ = row["CDI_Dermal_Child"] / RfD[metal]

            results.append([
                row["Sample"],
                metal,
                row["CDI_Dermal_Child"],
                HQ
            ])

    table = pd.DataFrame(results,
        columns=["Sample","Metal","CDI","HQ"])

    table.to_csv("Ajakanga_output\\Child_risk\\HQ_Dermal_Child.csv",index=False)

    return table
```

```python
def calculate_hq_inhalation(cdi_table, RfD):

    results = []

    for i,row in cdi_table.iterrows():

        metal = row["Metal"]

        if metal in RfD:

            HQ = row["CDI_Inhalation_Child"] / RfD[metal]

            results.append([
                row["Sample"],
                metal,
                row["CDI_Inhalation_Child"],
                HQ
            ])

    table = pd.DataFrame(results,
        columns=["Sample","Metal","CDI","HQ"])

    table.to_csv("Ajakanga_output\\Child_risk\\HQ_Inhalation_Child.csv",index=False)

    return table
```

```python
def calculate_cr_ingest_child(cdi_ingestion_table, CSF):

    results = []

    for i,row in cdi_ingestion_table.iterrows():

        metal = row["Metal"]

        if metal in CSF:

            CR = row["CDI_Ingestion_Child"] * CSF[metal]

            results.append([
                row["Sample"],
                metal,
                row["CDI_Ingestion_Child"],
                CR
            ])

    table = pd.DataFrame(results,
        columns=["Sample","Metal","CDI","Cancer_Risk"])

    table.to_csv("Ajakanga_output\\Child_risk\\CR_Ingestion_Child.csv",index=False)

    return table
```

```python
def calculate_cr_dermal_child(cdi_dermal_table, CSF):

    results = []

    for i,row in cdi_dermal_table.iterrows():

        metal = row["Metal"]

        if metal in CSF:

            CR = row["CDI_Dermal_Child"] * CSF[metal]

            if CR < 1e-6:
                classification = "Negligible"
                interpretation = "Insignificant cancer risk"

            elif CR <= 1e-4:
                classification = "Acceptable"
                interpretation = "Within acceptable cancer risk range"

            else:
                classification = "High Risk"
                interpretation = "Potential carcinogenic risk"

            results.append([
                row["Sample"],
                metal,
                row["CDI_Dermal_Child"],
                CSF[metal],
                CR,
                classification,
                interpretation
            ])

    cr_dermal_table = pd.DataFrame(
        results,
        columns=[
            "Sample",
            "Metal",
            "CDI_Dermal",
            "CSF",
            "Cancer_Risk_Dermal",
            "CR_Class",
            "CR_Interpretation"
        ]
    )

    cr_dermal_table.to_csv("Ajakanga_output\\Child_risk\\CR_Dermal_Child.csv", index=False)

    return cr_dermal_table
```

```python
def calculate_cr_inhalation_child(cdi_inhalation_table, CSF):

    results = []

    for i,row in cdi_inhalation_table.iterrows():

        metal = row["Metal"]

        if metal in CSF:

            CR = row["CDI_Inhalation_Child"] * CSF[metal]

            if CR < 1e-6:
                classification = "Negligible"
                interpretation = "Insignificant cancer risk"

            elif CR <= 1e-4:
                classification = "Acceptable"
                interpretation = "Within acceptable cancer risk range"

            else:
                classification = "High Risk"
                interpretation = "Potential carcinogenic risk"

            results.append([
                row["Sample"],
                metal,
                row["CDI_Inhalation_Child"],
                CSF[metal],
                CR,
                classification,
                interpretation
            ])

    cr_inhalation_table = pd.DataFrame(
        results,
        columns=[
            "Sample",
            "Metal",
            "CDI_Inhalation",
            "CSF",
            "Cancer_Risk_Inhalation",
            "CR_Class",
            "CR_Interpretation"
        ]
    )

    cr_inhalation_table.to_csv("Ajakanga_output\\Child_risk\\CR_Inhalation_Child.csv", index=False)

    return cr_inhalation_table
```

```python
import pandas as pd

# ====================================
# IMPORT DATA
# ====================================

hq_ing = pd.read_csv("Ajakanga_output\\Child_risk\\HQ_Ingestion_Child.csv")
hq_der = pd.read_csv("Ajakanga_output\\Child_risk\\HQ_Dermal_Child.csv")
hq_inh = pd.read_csv("Ajakanga_output\\Child_risk\\HQ_Inhalation_Child.csv")

cr_ing = pd.read_csv("Ajakanga_output\\Child_risk\\CR_Ingestion_Child.csv")
cr_der = pd.read_csv("Ajakanga_output\\Child_risk\\CR_Dermal_Child.csv")
cr_inh = pd.read_csv("Ajakanga_output\\Child_risk\\CR_Inhalation_Child.csv")

# ====================================
# HAZARD INDEX (HI) PER PATHWAY
# ====================================

def calculate_hi(df, pathway):

    hi = (
        df.groupby("Sample")["HQ"]
        .sum()
        .reset_index()
    )

    hi.rename(columns={"HQ": "HI"}, inplace=True)
    hi["Pathway"] = pathway

    hi.to_csv(f"Ajakanga_output\\Child_risk\\HI_{pathway}.csv", index=False)

    return hi


hi_ing = calculate_hi(hq_ing, "Ingestion")
hi_der = calculate_hi(hq_der, "Dermal")
hi_inh = calculate_hi(hq_inh, "Inhalation")

# ====================================
# TOTAL HAZARD INDEX (HI_total)
# ====================================

all_hi = pd.concat([hi_ing, hi_der, hi_inh])

hi_total = (
    all_hi.groupby("Sample")["HI"]
    .sum()
    .reset_index()
)

hi_total.rename(columns={"HI": "HI_Total"}, inplace=True)

hi_total.to_csv("Ajakanga_output\\Child_risk\\HI_Total.csv", index=False)

# ====================================
# TOTAL CANCER RISK PER PATHWAY
# ====================================

def calculate_tcr_pathway(df, pathway):

    tcr = (
        df.groupby("Sample")["Cancer_Risk"]
        .sum()
        .reset_index()
    )

    tcr.rename(columns={"Cancer_Risk": "TCR"}, inplace=True)
    tcr["Pathway"] = pathway

    tcr.to_csv(f"Ajakanga_output\\Child_risk\\TCR_{pathway}.csv", index=False)

    return tcr


tcr_ing = calculate_tcr_pathway(cr_ing, "Ingestion")
tcr_der = calculate_tcr_pathway(cr_der, "Dermal")
tcr_inh = calculate_tcr_pathway(cr_inh, "Inhalation")

# ====================================
# TOTAL CANCER RISK (ALL PATHWAYS)
# ====================================

all_tcr = pd.concat([tcr_ing, tcr_der, tcr_inh])

tcr_total = (
    all_tcr.groupby("Sample")["TCR"]
    .sum()
    .reset_index()
)

tcr_total.rename(columns={"TCR": "Total_Cancer_Risk"}, inplace=True)

tcr_total.to_csv("Ajakanga_output\\Child_risk\\TCR_Total.csv", index=False)

# ====================================
# OPTIONAL: MERGED RISK SUMMARY TABLE
# ====================================

summary = hi_total.merge(tcr_total, on="Sample")

summary.to_csv("Ajakanga_output\\Child_risk\\Risk_Assessment_Summary.csv", index=False)

print("All risk metrics successfully calculated and exported.")
```
